```python
#First method
# download the iMDb datasheet from kaggle here the link: https://www.kaggle.com/datasets/sankha1998/tmdb-top-10000-popular-movies-dataset
# Step 1: Import necessary libraries
import pandas as pd  # Library for data manipulation and analysis
import matplotlib.pyplot as plt  # Library for creating visualizations
from nltk.sentiment import SentimentIntensityAnalyzer  # Sentiment analysis library

# Step 2: Load the dataset
file_path = '/home/sohail-pc/Downloads/TMDb_updated.CSV' # replace this with your own file path
df = pd.read_csv(file_path)  # Load the dataset into a pandas DataFrame

# Step 3: Handle NaN values in the 'overview' column
df['overview'] = df['overview'].fillna('')  # Replace NaN values with an empty string

# Step 4: Set up sentiment analysis
sia = SentimentIntensityAnalyzer()  # Create a SentimentIntensityAnalyzer object

# Step 5: Define a function to get sentiment polarity
def get_sentiment_polarity(text):
    sentiment_score = sia.polarity_scores(text)
    return sentiment_score['compound']

# Step 6: Apply sentiment analysis to the 'overview' column and create a new 'Sentiment' column
df['Sentiment'] = df['overview'].apply(get_sentiment_polarity)

# Step 7: Sort the DataFrame by 'vote_count' in descending order and select the top 10 movies
df_top10 = df.sort_values(by='vote_count', ascending=False).head(10)

# Step 8: Plot a bar chart to visualize sentiment scores for the top 10 movies
plt.figure(figsize=(12, 6))  # Set the size of the plot
plt.bar(df_top10['title'], df_top10['Sentiment'], color='skyblue')  # Create a bar chart
plt.title('Sentiment Scores for Top 10 Movies (Based on Vote Count)')  # Set the title
plt.xlabel('Movie Titles')  # Label for the x-axis
plt.ylabel('Sentiment Score')  # Label for the y-axis
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout for better presentation
plt.show()  # Display the plot
```

```python

# Step 1: Import necessary libraries
import pandas as pd  # Library for data manipulation and analysis
from nltk.sentiment import SentimentIntensityAnalyzer  # Library for sentiment analysis

# Step 2: Load the dataset
file_path = '/home/sohail-pc/Downloads/TMDb_updated.CSV'  # Specify the path to the movie dataset CSV file
df = pd.read_csv(file_path)  # Load the dataset into a pandas DataFrame

# Step 3: Handle NaN values in the 'overview' column
df['overview'] = df['overview'].fillna('')  # Replace NaN values in the 'overview' column with an empty string

# Step 4: Set up sentiment analysis
sia = SentimentIntensityAnalyzer()  # Create a SentimentIntensityAnalyzer object

# Step 5: Define a function to get sentiment polarity
def get_sentiment_polarity(text):
    sentiment_score = sia.polarity_scores(text)
    return sentiment_score['compound']

# Step 6: Apply sentiment analysis to the 'overview' column
df['Sentiment'] = df['overview'].apply(get_sentiment_polarity)
# Calculate sentiment polarity scores for each 'overview' text and create a new 'Sentiment' column in the DataFrame

# Step 7: Sort the DataFrame by 'vote_count' in descending order and select the top 10 movies
df_top10 = df.sort_values(by='vote_count', ascending=False).head(10)

# Step 8: Display the top 10 movies with their vote counts and sentiment scores
for index, row in df_top10.iterrows():
    print(f"Title: {row['title']}, Vote Count: {row['vote_count']}, Sentiment: {row['Sentiment']:.2f}")
```
